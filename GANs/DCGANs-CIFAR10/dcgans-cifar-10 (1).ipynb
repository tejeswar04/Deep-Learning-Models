```python
#Define discriminator
# define the standalone discriminator model
def discriminator(in_shape=(32,32,3)):
    model=Sequential()
    #Normal CNN
    model.add(Conv2D(64,(3,3),padding='same',input_shape=in_shape))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2D(128,(3,3),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2D(128,(3,3),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2D(256,(3,3),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Flatten())
    model.add(Dropout(0.4))
    model.add(Dense(1,activation='sigmoid'))
    opt = Adam(lr=0.0002,beta_1=0.5)
    model.compile(loss='binary_crossentropy',optimizer=opt,metrics=['accuracy'])
    return model
```

```python
#Changing the pixel values to float and scalling from [0,255] to [-1,1]
X=TrainX.astype('float32')
X=(X-127.5)/127.5
```

```python
def Gen_Real_Samples(dataset,n_samples):
    ix=np.random.randint(0,dataset.shape[0],n_samples)
    X=dataset[ix]
    y=np.ones((n_samples,1))
    return X,y
```

```python
def Gen_Fake_Samples(n_samples):
    X=np.random.rand(32*32*3*n_samples)
    #Changing them from [0,1] to [-1,1]
    X=-1+2*X
    X=X.reshape((n_samples,32,32,3))
    y=np.zeros((n_samples,1))
    return X,y
```

```python
def Train_Discriminator(model,dataset,epoch=20,n_batch=128):
    half_batch=int(n_batch/2)
    for i in range(epoch):
        X_real,Y_real=Gen_Real_Samples(dataset,half_batch)
        #It is told in papers that It is stable to train 1's together and 0's together
        _,real_acc=model.train_on_batch(X_real,Y_real)
        X_fake,Y_fake=Gen_Fake_Samples(half_batch)
        _,fake_acc=model.train_on_batch(X_fake,Y_fake)
        print('>%d real=%.0f%% fake=%.0f%%' % (i+1, real_acc*100, fake_acc*100))
```

```python
model=discriminator()
Train_Discriminator(model,X)
```

```python
def Generator(latent_dim):
    model=Sequential()
    n_nodes=256*4*4
    model.add(Dense(n_nodes,input_dim=latent_dim))
    model.add(LeakyReLU(0.2))
    model.add(Reshape((4,4,256)))
    model.add(Conv2DTranspose(128,(4,4),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2DTranspose(128,(4,4),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2DTranspose(128,(4,4),strides=(2,2),padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Conv2D(3,(3,3),activation='tanh',padding='same'))
    return model
```

```python
latent_dim=100
model_gen=Generator(latent_dim)
model_gen.summary()
```

```python
def Gen_Latent_Points(latent_dim,n_samples):
    x_input=randn(latent_dim*n_samples)
    x_input=x_input.reshape(n_samples,latent_dim)
    return x_input
```

```python
def Gen_Fake_Samples_G(g_model,latent_dim,n_samples):
    x_input=Gen_Latent_Points(latent_dim,n_samples)
    X=g_model.predict(x_input)
    Y=zeros((n_samples,1))
    return X,Y
```

```python
n_samples=49
latent_dim=100
X_g,_=Gen_Fake_Samples_G(model_gen,latent_dim,n_samples)
#Scaling from [-1,1] to [0,1]
X_g=(X_g+1)/2
for i in range(49):
    pyplot.subplot(7,7,1+i)
    pyplot.axis('off')
    pyplot.imshow(X_g[i])
pyplot.show()
```

```python
def GAN(g_model,d_model):
    d_model.trainable=False
    model=Sequential()
    model.add(g_model)
    model.add(d_model)
    opt=Adam(lr=0.0002,beta_1=0.5)
    model.compile(loss='binary_crossentropy',optimizer=opt)
    return model
```

```python
d_model=discriminator()
g_model=Generator(latent_dim)
GAN_sample=GAN(g_model,d_model)
GAN_sample.summary()
```

```python
def Train_GAN(gan_model,latent_dim,epoch=200,n_batch=128):
    for i in range(epoch):
        x_gan=Gen_Latent_Points(latent_dim,n_batch)
        y_gan=ones((n_batch,1))
        gan_model.train_on_batch(x_gan,y_gan)
```

```python
def save_plot(examples, epoch, n=7):
    examples = (examples + 1) / 2.0
    for i in range(n * n):
        pyplot.subplot(n, n, 1 + i)
        pyplot.axis('off')
        pyplot.imshow(examples[i])
        filename = 'generated_plot_e%03d.png' % (epoch+1)
        pyplot.savefig(filename)
        pyplot.close()
```

```python
def summarize_performance(epoch, g_model, d_model, dataset, latent_dim, n_samples=150):
    X_real, y_real = Gen_Real_Samples(dataset, n_samples)
    _, acc_real = d_model.evaluate(X_real, y_real, verbose=0)
    x_fake, y_fake = Gen_Fake_Samples_G(g_model, latent_dim, n_samples)
    _, acc_fake = d_model.evaluate(x_fake, y_fake, verbose=0)
    print('>Accuracy real: %.0f%%, fake: %.0f%%' % (acc_real*100, acc_fake*100))
    save_plot(x_fake, epoch)
    filename = 'generator_model_%03d.h5' % (epoch+1)
    g_model.save(filename)
```

```python
def Train(g_model,d_model,gan_model,dataset,latent_dim,epoch=50,n_batch=128):
    bat_per_epo=int(dataset.shape[0]/n_batch)
    half_batch=int(n_batch/2)
    for i in range(epoch):
        for j in range(bat_per_epo):
            X_real,Y_real=Gen_Real_Samples(dataset,half_batch)
            d_loss1,_=d_model.train_on_batch(X_real,Y_real)
            X_fake,Y_fake=Gen_Fake_Samples_G(g_model,latent_dim,half_batch)
            d_loss2,_=d_model.train_on_batch(X_fake,Y_fake)
            X_gan=Gen_Latent_Points(latent_dim,n_batch)
            y_gan=ones((n_batch,1))
            g_loss=gan_model.train_on_batch(X_gan,y_gan)
            print('>%d, %d/%d, d1=%.3f, d2=%.3f g=%.3f' %
            (i+1, j+1, bat_per_epo, d_loss1, d_loss2, g_loss))
            if (i+1) % 10 == 0:
                summarize_performance(i, g_model, d_model, dataset, latent_dim)
```

```python
latent_dim=100
d_model=discriminator()
g_model=Generator(latent_dim)
gan_model=GAN(g_model,d_model)
dataset=X
Train(g_model,d_model,gan_model,dataset,latent_dim)
```
